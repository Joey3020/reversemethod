```python
import pandas as pd
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F 
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import time
import math

# use GPU??
USE_GPU = True
if USE_GPU and torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')

    
print(device)


LEARNING_RATE = 1e-4
BATCH_SIZE = 64
EPOCH_NUM = 100
```

```python
class DataLoader:
    def __init__(self):
        self.current_batch = 0
        self.batch_size = 0
        self.batch_num = 0
        
        # load training data
        input1_pd = pd.read_excel(open('train_data.xlsx', 'rb'), sheet_name = 'Sheet1', header = None)
        input_ten = torch.from_numpy( np.array(input1_pd) ).float()
        self.input = input_ten.reshape(-1,1,32,32)
        
        label_ab_pd = pd.read_excel(open('train_data.xlsx', 'rb'), sheet_name = 'Sheet2', header = None)
        self.label_ab = torch.from_numpy( np.array(label_ab_pd) ).float()
        
        label_mode_pd = pd.read_excel(open('train_data.xlsx', 'rb'), sheet_name = 'Sheet3', header = None)
        self.label_mode = torch.from_numpy( np.array(label_mode_pd) )
        
        label_phase_pd = pd.read_excel(open('train_data.xlsx', 'rb'), sheet_name = 'Sheet4', header = None)
        self.label_phase = torch.from_numpy( np.array(label_phase_pd) ).float()
        
        self.data_size = len(label_ab_pd.index)
        print(self.data_size)
        
    def set_batch_size(self, size):
        self.batch_size = size

    def get_batch_num(self):
        self.batch_num = math.ceil(self.data_size / self.batch_size)
        return self.batch_num

    def get_batch(self):
        """
        전체 trainig 데이터에서 set_batch_size 에서 정한 batch size만큼만 데이터를 리턴합니다.
        이 떄 어디까지 리턴을 했는지를 저장하고 있다가 이전에 리턴한 데이터 다음 데이터부터 리턴을 합니다.
        나머지 데이터가 batch size보다 작을경우에는 남은 데이터만 리턴하고 어디까지 리턴했는지를 초기화 합니다.

        :return: batch
        """
        length = self.batch_size
        current_batch = self.current_batch
        
        if current_batch + length <= self.data_size :
            data_return = self.input[ current_batch : current_batch + length ]
            #label_return = self.y_onehot[ current_batch : current_batch + length ] ##
            
            #label_return = torch.cat([self.label_ab[ current_batch : current_batch + length ], self.label_mode[ current_batch : current_batch + length ], self.label_phase[ current_batch : current_batch + length ]], dim=1)
            self.current_batch += length
            return data_return, self.label_ab[ current_batch : current_batch + length ], self.label_mode[ current_batch : current_batch + length ], self.label_phase[ current_batch : current_batch + length ]
        else:
            data_return = self.input[ current_batch : -1 ]
            #label_return = self.y_onehot[ current_batch : -1 ]
            #label_return = torch.cat([self.label_ab[ current_batch : -1 ], self.label_mode[ current_batch : -1 ], self.label_phase[ current_batch : -1 ]], dim=1)
            self.current_batch = 0
            return data_return, self.label_ab[ current_batch : -1], self.label_mode[ current_batch : -1 ], self.label_phase[ current_batch :-1 ]
        #return data_return, label_return
        
    def reset(self): #어디까지 리턴했는지 초기화
        self.current_batch = 0
```

```python
loader = DataLoader()
loader.set_batch_size(BATCH_SIZE)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.cnn = nn.Sequential(
            # 3 x 32 x 32
            nn.Conv2d(1, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            
            # 64 x 16 x 16
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(128),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            
            # 128 x 8 x 8
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(256),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            
            
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            #512 x 2 x 2
        )
        self.fc_ab = nn.Sequential(
            nn.Linear(512*2*2, 1024),
            nn.ReLU(),
            nn.Dropout(0.5),

            nn.Linear(1024, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            #nn.Linear(1024, 1)
            nn.Linear(256, 2)
        )
        
        self.fc_mode = nn.Sequential(
            nn.Linear(512*2*2, 1024),
            nn.ReLU(),
            nn.Dropout(0.5),

            nn.Linear(1024, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            #nn.Linear(1024, 1)
            nn.Linear(256, 5)
        )
        
        self.fc_phase = nn.Sequential(
            nn.Linear(512*2*2, 1024),
            nn.ReLU(),
            nn.Dropout(0.5),

            nn.Linear(1024, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            #nn.Linear(1024, 1)
            nn.Linear(256, 1)
        )
        
    def forward(self, x):
        output = self.cnn(x)
        output = output.view(output.size(0), -1)
        
        output_ab = self.fc_ab(output)
        
        output_mode = self.fc_mode(output)
        #output_mode = F.log_softmax(output_mode)
        
        output_phase = self.fc_phase(output)
        
        
        return output_ab, output_mode, output_phase
```

```python
class Model:

    def __init__(self):
        self.net = Net()
        self.criterion1 = nn.MSELoss()
        self.criterion2 = nn.CrossEntropyLoss()
        self.criterion3 = nn.MSELoss()
        self.ismodelsaved = False

    def set_learning_rate(self, learning_rate):
        self.learning_rate = learning_rate
        self.optimizer = optim.Adam(self.net.parameters(), lr=learning_rate, betas=(0.9, 0.999), eps=1e-08, weight_decay=0, amsgrad=False)

    def train_step(self, x, ab, mode, phase):
        #print(x), print(y)
        self.optimizer.zero_grad()
        infer_ab, infer_mode, infer_phase = self.net(x)
        
        loss1 = self.criterion1(infer_ab, ab)
        mode = mode.squeeze(1)
        loss2 = self.criterion2(infer_mode, mode)
        loss3 = self.criterion3(infer_phase, phase)
        loss = loss1 + loss2 + loss3
        loss.backward()
        self.optimizer.step()
        self.ismodelsaved = False # model을 saved 하고 load하고 다시 training 했을 때 새로운 모델은 저장이 안된 것으로 표시
        return loss
    
    def inference(self, x):
        return self.net(x)

    def save(self):
        torch.save(self.net.state_dict(), 'model.pt')
        self.ismodelsaved = True
        
    def restore(self):
        if self.ismodelsaved == True:
            # model = TheModelClass(*args, **kwargs)
            self.net.load_state_dict(torch.load( 'model.pt'))
            self.net.eval()
        
        return self.ismodelsaved
```

```python
model = Model()
# model.net = model.net.float()
model.set_learning_rate(LEARNING_RATE)
model.net.to(device)
```

```python
#training step
input_val = pd.read_excel(open('validation_data.xlsx', 'rb'), sheet_name = 'Sheet1', header = None)
input_val = torch.from_numpy( np.array(input_val) ).float()
input_val = input_val.reshape(-1,1,32,32)
        
label_val_ab = pd.read_excel(open('validation_data.xlsx', 'rb'), sheet_name = 'Sheet2', header = None)
label_val_ab = torch.from_numpy( np.array(label_val_ab) ).float()
        
label_val_mode = pd.read_excel(open('validation_data.xlsx', 'rb'), sheet_name = 'Sheet3', header = None)
label_val_mode = torch.from_numpy( np.array(label_val_mode) )
        
label_val_phase = pd.read_excel(open('validation_data.xlsx', 'rb'), sheet_name = 'Sheet4', header = None)
label_val_phase = torch.from_numpy( np.array(label_val_phase) ).float()
    

batch_num = loader.get_batch_num()
cost_list = []
val_list_ab = []
val_list_mode = []
val_list_phase = []
for epoch_idx in range(EPOCH_NUM):
    cumcost = 0.0
    for idx in range(batch_num):
        x, ab, mode, phase = loader.get_batch()
        
        x = x.to(device)
        ab = ab.to(device)
        mode = mode.to(device)
        phase = phase.to(device)
        cumcost += model.train_step(x, ab, mode, phase)
    cumcost /= batch_num
    
    cost_list.append(cumcost)

    with torch.no_grad():
        val_size = len(label_val_ab)
        
        input_val = input_val.to(device)
        label_val_ab = label_val_ab.to(device)
        label_val_mode = label_val_mode.to(device)
        label_val_phase = label_val_phase.to(device)
        
        pred_ab, pred_mode, pred_phase = model.inference(input_val)
        
        error_val_ab = pred_ab - label_val_ab
        relative_error_val_ab = error_val_ab / label_val_ab
        
        
        _, pred_val_mode = pred_mode.max(1)
        correct_mode_val = 0
        for i in range(val_size):
            if pred_val_mode[i] == label_val_mode[i] :
                correct_mode_val+=1
        accuracy_val_mode = correct_mode_val/val_size
        
        error_val_phase = pred_phase - label_val_phase
        relative_error_val_phase = error_val_phase / label_val_phase
        
    val_list_ab.append(torch.mean(torch.abs(relative_error_val_ab)))
    val_list_mode.append(1-accuracy_val_mode)
    val_list_phase.append(torch.mean(torch.abs(relative_error_val_phase)))
    
    if epoch_idx % 1 == 0:
        
        print('epoch num : ', epoch_idx, 'loss : ', cumcost)
        print('error for ab :', torch.mean(torch.abs(relative_error_val_ab)))
        print('accuracy for mode :', accuracy_val_mode)
        print('error for phase :', torch.mean(torch.abs(relative_error_val_phase)))
        print('----------------------------------------------------------------------------')
    cost_list.append(cumcost)
    
print('Finished Training')
```

```python
import matplotlib.pyplot as plt

plt.plot(cost_list)
plt.plot(val_list_ab)
plt.plot(val_list_mode)
plt.plot(val_list_phase)
plt.show()

model.save()
```
